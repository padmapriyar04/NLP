```python
import spacy
import pandas as pd
```

```python
data = pd.read_csv("bbc_news.csv")
data.head(5)
```

```python
nlp = spacy.load("en_core_web_sm")
dataTitle = pd.DataFrame(data['title'])
pos_doc_title = dataTitle.apply(lambda x: [nlp(text) for text in x])
pos_doc_title
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
#removal of stop words:
dataTitle['title_lower'] = dataTitle['title'].str.lower()
en_stopwords = stopwords.words('english')
dataTitle['title_nostop'] = dataTitle['title_lower'].apply(lambda x : ' '.join([word for word in x if word not in en_stopwords]))
dataTitle['title_nostop']
```

```python
#removal of punctuations:
dataTitle['title_nopunct'] =dataTitle.apply(lambda x : re.sub(r"([^\w\s])","",x['title_nostop']),axis=1)
dataTitle['title_nopunct']
```

```python
#tokenise the data Titles:

dataTitle['tokenised'] = dataTitle.apply(lambda x : word_tokenize(x['title_nopunct']),axis=1)
dataTitle['tokenised']
```

```python
nltk.download('wordnet')
wdlem = WordNetLemmatizer()
dataTitle['title_lemmatised'] = dataTitle['tokenised'].apply(lambda x : [wdlem.lemmatize(word) for word in x])
dataTitle['title_lemmatised']
```

```python
title_final = sum(dataTitle['title_lemmatised'],[])
unigrams = pd.DataFrame(nltk.ngrams(title_final,1))
unigrams.value_counts()
```
