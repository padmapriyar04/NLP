```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
import io
```

```python
data = pd.read_csv(io.BytesIO(uploaded['tripadvisor_hotel_reviews.csv']))
```

```python
data.sample(3)
```

```python
data["Review"][92]
```

```python
data["review_lc"] = data["Review"].str.lower()
data.head()
```

```python
nltk.download('stopwords')
en_stopwords = stopwords.words("english")
en_stopwords.remove('not')
```

```python
data['reviews_nostop'] = data['review_lc'].apply(lambda x: ' '.join([word for word in x.split() if word not in en_stopwords]))
data['reviews_nostop']
```

```python
data['reviews_no_punct'] = data.apply(lambda x : re.sub(r"[*]",'star',x['reviews_nostop']),axis=1)
```

```python
data['processed'] = data.apply(lambda x : re.sub(r"([^\w\s])","",x['reviews_no_punct']),axis=1)
data['processed']
```

```python
nltk.download('punkt_tab')
data['tokenised'] = data.apply(lambda x : word_tokenize(x['processed']),axis=1)
data.head(3)
```

```python
ps = PorterStemmer()
data['Stemmed'] = data['tokenised'].apply(lambda x : [ps.stem(word) for word in x])
data.head()
```

```python
nltk.download('wordnet')
wdLem = WordNetLemmatizer()
data['Lemmatised'] = data['tokenised'].apply(lambda x : [wdLem.lemmatize(word) for word in x])
data['Lemmatised']
```

```python
tokens_cleaned = sum(data['Lemmatised'],[])
unigrams = (pd.Series(nltk.ngrams(tokens_cleaned,1)))
bigrams = (pd.Series(nltk.ngrams(tokens_cleaned,2)))
bigrams.value_counts()
```
